```python
#PLot 1*4 subplots
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
```

```python
pathimage = 'imageViz'
```

```python
images = os.listdir(pathimage)
images = [img for img in images if img.endswith('.png')]
images = sorted(images)
print(images)
```

```python
for i in range(4):
    plt.subplot(1, 4, i+1)
    img = cv2.imread(os.path.join(pathimage, images[i]))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)


    fig = plt.gcf()
    fig.set_size_inches(18, 18)

    plt.axis('off')

    if i < 3:
        plt.title('Keine Wartelinie'.format(i+1))
    else:
        plt.title('Wartelinie'.format(i+1))

plt.savefig(os.path.join(pathimage, 'imageViz.png'))
plt.show()
```
